```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

1. Read datasets, and preprocessing

```python
data = pd.read_csv('Dataset/Iris.csv')

standartScaler = StandardScaler()
x = data.drop(['Species'], axis=1)
y = data['Species']
x = standartScaler.fit_transform(x)
scaler = LabelEncoder()
y = scaler.fit_transform(y)

intercept = np.ones((x.shape[0], 1))
x = np.concatenate((intercept, x), axis=1)
print(x)
# print(x)

# print(y)

print(data.shape)
```

2. Split datasets into training and test sets

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.5, random_state=42)

print(x_train)

print(y_train)
```

3. Using softmax to solve this problem

```python
N = x_train.shape[0]
k = 3 
d = data.shape[1]
def compute_one_hot(y_train, k):

    one_hot = np.zeros((len(y_train), k))
    one_hot[np.arange(len(y_train)), y_train] = 1
    return one_hot

y_one_hot = compute_one_hot(y_train, k)

#print(y_one_hot)


print(d)
```

```python
weights = 0.01 * np.random.randn(d, k)
print('theta:\n', weights)
print(x_train.shape)
print(weights.shape)
```

```python
learing_rate = 0.01
epochs = 1000
losses = []

for epoch in range(epochs):

    # Probability of each class
    z = x_train.dot(weights)
    max = np.max(z, axis = 1, keepdims=True)
    exp = np.exp(z - max)
    probability = exp / np.sum(exp,axis = 1,keepdims=True)

    # Loss 

    loss = -(y_one_hot * np.log(probability)).sum()
    losses.append(loss)

    # Gradient

    gradient = probability - y_one_hot

    dTheta = x_train.T.dot(gradient)

    # Update weights
    weights -= learing_rate * dTheta
    
```

```python
prediction = x_test.dot(weights)

print(prediction)
predicted_class = np.argmax(prediction, axis=1)

print("Predicted class: ", predicted_class)

print("Actual class: ", y_test)

accuracy = np.mean(predicted_class == y_test) * 100

print("Accuracy: ", accuracy)
```
